```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('data.csv')
df
```

```python
import math
valueBedrooms = math.floor(df['bedrooms'].median())
df['bedrooms'] = df['bedrooms'].fillna(valueBedrooms)
print(df['bedrooms'])
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df['price'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
dfHiring = pd.read_csv('hiring.csv')
dfHiring
```

```python
dfHiring['experience'] = dfHiring['experience'].apply(lambda x: w2n.word_to_num(x) if isinstance(x,str) else x)
dfHiring['experience'] = dfHiring['experience'].fillna(0)
dfHiring['experience']
```

```python
valueNAN = math.floor(dfHiring['test_score(out of 10)'].median())
dfHiring['test_score(out of 10)'] = dfHiring['test_score(out of 10)'].fillna(valueNAN)
dfHiring['test_score(out of 10)']
```

```python
dfHiring
```

```python
reghiring = linear_model.LinearRegression()
reghiring.fit(dfHiring[['experience','test_score(out of 10)','interview_score(out of 10)']],dfHiring['salary($)'])
```

```python
reghiring.coef_
```

```python
reghiring.intercept_
```

```python
reghiring.predict([[2,9,6]])
```
