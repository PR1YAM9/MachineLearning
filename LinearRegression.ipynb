```python
df = pd.read_csv('data.csv.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
dfCan = pd.read_csv('canada_per_capita_income.csv')
dfCan.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.plot(dfCan[['year']], dfCan["per capita income (US$)"])
```

```python
regCan = linear_model.LinearRegression()
regCan.fit(dfCan[['year']],dfCan["per capita income (US$)"])
```

```python
regCan.coef_
```

```python
regCan.intercept_
```

```python
regCan.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('years')
plt.ylabel('per capita income (US$)')
plt.plot(dfCan["per capita income (US$)"], regCan.predict(dfCan[['year']]), color = 'red')
```
